2. Find out the number, of the highest quantity of orders for a product, then multiply by its Order ID

```sql
SELECT sum(quantity) as total, productid FROM orderdetails
group by productid
ORDER BY total DESC;
```
Order details have a many to 1 relationship with orders - each record is an order id, productid, and quant probably some other stuff but i cant be bothered to do a select *.

From this we can use the above logic to group by product id and take a sum of the totals, this gave 458 quantity of id 31
